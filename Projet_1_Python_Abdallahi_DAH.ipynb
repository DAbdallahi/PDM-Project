# Part 1: Dataset Exploration

**Objective: Open and explore the provided datasets.**

Examine the structure of the SQLite database containing wellbore data (wells_data.db).
Explore the Excel file with daily and monthly production data (multi-tab format).
Gain a clear understanding of the data contents and structure.

```python
# Load data from the SQLite database
conn_sqlite = sqlite3.connect('wells_data.db')
wells_data = pd.read_sql_query("SELECT * FROM wells_data", conn_sqlite)

# Load Excel data with multiple sheets
excel_file = pd.ExcelFile('Volve_production_data.xlsx')
daily_production = pd.read_excel(excel_file, sheet_name='Daily Production Data')
monthly_production = pd.read_excel(excel_file, sheet_name='Monthly Production Data')
```

# Part 3: Data Analysis and Visualization
Objective: Analyze and visualize the wellbore data.



### 1.1 & 1.2. Create a scatter plot of:
 - Bottom hole temperature (vertical axis) vs. total depth (horizontal axis), both with and without excluding 0° temperatures.
 - Drilling days (vertical axis) vs. total depth (horizontal axis), filtering rows with drilling days between 0 and 800.

```python
# Set the seaborn style for better aesthetics
sns.set(style='whitegrid')

# Create a figure with 3 subplots (1 row, 3 columns)
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Scatter plot without removing 0°C temperature values
sns.scatterplot(data=wells_data, x="wlbTotalDepth", y="wlbBottomHoleTemperature", ax=axes[0], color='orange')
axes[0].set_title("Bottom Hole Temperature vs. Total Depth (with 0°C)", fontsize=14)
axes[0].set_xlabel("Total Depth", fontsize=12)
axes[0].set_ylabel("Bottom Hole Temperature (°C)", fontsize=12)

# Scatter plot removing 0°C temperature values
filtered_data = wells_data[wells_data["wlbBottomHoleTemperature"] > 0]
sns.scatterplot(data=filtered_data, x="wlbTotalDepth", y="wlbBottomHoleTemperature", ax=axes[1], color='blue')
axes[1].set_title("Bottom Hole Temperature vs. Total Depth (without 0°C)", fontsize=14)
axes[1].set_xlabel("Total Depth", fontsize=12)
axes[1].set_ylabel("Bottom Hole Temperature (°C)", fontsize=12)

# Filter data with `query` to keep only drilling days in the desired range
drilling_days_data = wells_data.query("0 < wlbDrillingDays < 800")

# Scatter plot of drilling days vs. total depth
sns.scatterplot(data=drilling_days_data, x="wlbTotalDepth", y="wlbDrillingDays", ax=axes[2], color='green')
axes[2].set_title("Drilling Days vs. Total Depth (0 < wlbDrillingDays < 800)", fontsize=14)
axes[2].set_xlabel("Total Depth", fontsize=12)
axes[2].set_ylabel("Drilling Days", fontsize=12)

# Adjust the layout
plt.tight_layout()
plt.show()
```

### 1.3. Calculate and display:
The mean total depth grouped by "Age At TD" as a sorted table

.

```python
# Calculate the average total depth by 'Age At Td' and sort by mean total depth
mean_depth_by_age = wells_data.groupby("wlbAgeAtTd")["wlbTotalDepth"].mean().reset_index()
mean_depth_by_age = mean_depth_by_age.rename(columns={"wlbTotalDepth": "Mean Total Depth"})
mean_depth_by_age = mean_depth_by_age.sort_values(by="Mean Total Depth", ascending=True)

# Display the table
mean_depth_by_age.head()
```

# Part 4: Production Data Analysis

### 4.1 Monthly Production Trends:
Analyze and visualize monthly production data (Oil, Gas, and Water) for NPDCode 7405 to identify trends over time.

```python
# Filter the data for the NPD code 7405
volve_data = monthly_production[monthly_production['NPDCode'] == 7405]
```

```python
# Create a 'Month-Year' column by concatenating 'Year' and 'Month'
volve_data.loc[:, 'Month-Year'] = volve_data['Year'].astype(str) + '-' + volve_data['Month'].astype(str).str.zfill(2)
```

```python
# Convert 'Month-Year' to datetime format for chronological sorting
volve_data.loc[:, 'Month-Year'] = pd.to_datetime(volve_data['Month-Year'], format='%Y-%m')

# Create a figure with 3 subplots (1 row, 3 columns)
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Plot oil production
sns.lineplot(data=volve_data, x='Month-Year', y='Oil (Sm3)', ax=axes[0], color='brown', marker='o')
axes[0].set_title('Oil Production', fontsize=16)
axes[0].set_xlabel('Month-Year', fontsize=14)
axes[0].set_ylabel('Production (m³)', fontsize=14)
axes[0].tick_params(axis='x', rotation=30)
axes[0].grid(False)
axes[0].spines['top'].set_visible(False)
axes[0].spines['right'].set_visible(False)

# Plot gas production
sns.lineplot(data=volve_data, x='Month-Year', y='Gas (Sm3)', ax=axes[1], color='blue', marker='o')
axes[1].set_title('Gas Production', fontsize=16)
axes[1].set_xlabel('Month-Year', fontsize=14)
axes[1].set_ylabel('Production (m³)', fontsize=14)
axes[1].tick_params(axis='x', rotation=30)
axes[1].grid(False)
axes[1].spines['top'].set_visible(False)
axes[1].spines['right'].set_visible(False)

# Plot water production
sns.lineplot(data=volve_data, x='Month-Year', y='Water (Sm3)', ax=axes[2], color='green', marker='o')
axes[2].set_title('Water Production', fontsize=16)
axes[2].set_xlabel('Month-Year', fontsize=14)
axes[2].set_ylabel('Production (m³)', fontsize=14)
axes[2].tick_params(axis='x', rotation=30)
axes[2].grid(False)
axes[2].spines['top'].set_visible(False)
axes[2].spines['right'].set_visible(False)

# Adjust layout
plt.tight_layout()
plt.show()
```

### Total Oil Production:

# Part 5: Merged Data Analysis
In Part 5, we will merge the well data and monthly production data to create a new dataset. By focusing on key columns, we'll analyze the relationship between total depth and oil production using a scatter plot to check for any correlation between the two.

### 5.1 Data Preparation:
Extract relevant columns from both datasets and aggregate monthly production data by a common identifier.

.

```python
# Create partial_wells_data by keeping wlbTotalDepth and the common column
partial_wells_data = wells_data[['wlbWellboreName', 'wlbTotalDepth']]

# Create partial_mpd with the columns Oil, Gas, Water, and the common column,and aggregate data by summing grouped by the common column
partial_mpd = monthly_production[['Wellbore name', 'Oil (Sm3)', 'Gas (Sm3)', 'Water (Sm3)']]
partial_mpd = monthly_production.groupby('Wellbore name')[['Oil (Sm3)', 'Gas (Sm3)', 'Water (Sm3)']].sum().reset_index()
partial_mpd = partial_mpd.rename(columns={'Wellbore name': 'wlbWellboreName'})
```

```python
len(wells_data.wlbWellboreName.unique())
```

### 5.2. Data Integration:
Merge the prepared datasets into a unified merged_data DataFrame for analysis.

```python
merged_data = pd.merge(partial_wells_data, partial_mpd, on='wlbWellboreName', how='inner')
merged_data
```

### 5.3. Scatter Plot Analysis:
Visualize the relationship between oil production (vertical axis) and total depth (horizontal axis) to assess potential correlation.

```python
# Set the seaborn style for an elegant appearance
sns.set(style="whitegrid")

# Create a scatter plot for oil production versus total depth
plt.figure(figsize=(10, 4))
sns.scatterplot(data=merged_data, x="wlbTotalDepth", y="Oil (Sm3)", marker='o')

# Add labels and title
plt.xlabel("Total Depth (m)", fontsize=12)
plt.ylabel("Oil Production (Sm³)", fontsize=12)
plt.title("Oil Production vs. Total Depth", fontsize=14)

# Display the plot
plt.grid(True)
plt.show()
```

```python
# Calculate the correlation between oil production and total depth
correlation = merged_data['wlbTotalDepth'].corr(merged_data['Oil (Sm3)'])
print(f"Correlation between Total Depth and Oil Production: {correlation}")
```
